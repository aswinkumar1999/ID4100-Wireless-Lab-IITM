# Experiment 1

## Question 1 

Verify mean and standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu=0
sigma=0.1
samples=np.random.normal(mu,sigma,10000)
tolerance=1e-4
```

```python
# Verify mean
mu_loop=np.sum(samples)/len(samples)
print("Mean via for loop {}".format(mu_loop))

mu_np=np.mean(samples)
print(f"Mean via for numpy {mu_np}")

print("Verify with for loop {}".format(np.abs(mu_loop-mu<tolerance)))

print("Verify with np {}".format(np.abs(mu_np-mu<tolerance)))
```

```python
# Verify sigma
sigma_loop=np.sqrt(np.sum((samples-mu_loop)**2)/len(samples))
print("Sigma via for loop {}".format(sigma_loop))

sigma_np=np.std(samples)
print(f"Sigma via for numpy {sigma_np}")

print("Verify with loop {}".format(np.abs(sigma_loop-sigma<tolerance)))

print("Verify with np {}".format(np.abs(sigma_np-sigma<tolerance)))
```

```python
# Shift
k=10
mean_shifted=np.sum(samples-k)/len(samples)+k
print("Mean via shift {}".format(mean_shifted))
print("Verify with shift for mean {}".format(np.abs(mean_shifted-mu<tolerance*10)))

sigma_shifted=np.sqrt(np.sum((samples-k)**2)/len(samples)-(mean_shifted-k)**2)

print("Sigma via shift {}".format(sigma_shifted))
print("Verify with shift for mean {}".format(np.abs(sigma_shifted-sigma<tolerance)))
```
